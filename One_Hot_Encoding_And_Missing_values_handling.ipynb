```python
import pandas as pd
import numpy as np

#read data

data = pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
# labels each variable has

for col in data.columns:
    print(col, ':' , data[col].nunique(), 'labels')
```

```python
## one hot encoding

pd.get_dummies(data, drop_first=True).shape
```

```python
## top 10 most frequent categories for the variable x2

data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
## List with most frequent categories of the variable

top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# making 10 binary variables

for label in top_10:
    data[label] = np.where(data['X2']==label,1,0)
    
data[['X2']+top_10].head(40)
```

```python
def one_hot_top_x(df, variable, top_x_labels):
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(data[variable]==label,1,0)
        
data = pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])

one_hot_top_x(data,'X2',top_10)
data.head()
```

### Types of Encoding

1. Nominal Encoding \
    a. One hot encoding - use pd.get_dummies() \
    b. One hot encoding with many categories - sort variables first and take first 10 or 20 categories to get the encoding \
    c. Mean Encoding - Lets assume we have data like number of ordinal features and target variable is binary 1 or 0. We can calculate mean of target variable with respect to particular category of that feature. Assign means to feature according to the calculated mean value. 


2. Ordinal encoding \
    a. Label encoding - Use numbers as 1,2,3,4, etc. or use sklearn.label_encoder() \
    b. Target guided Ordinal encoding - Lets assume we have data like number of ordinal features and target variable is binary 1 or 0. We can calculate mean of target variable with respect to particular category of that feature. Assign ranks to feature according to the calculated mean value. That will be the value of that category for encoding.

#### Count or frequency encoding for High cardinality features

Replace each label of the categorical variable by the count \
This is the amount of times each label appears in the dataset. \
This is the percentage of observations within that category

```python
df = pd.read_csv('train.csv',usecols=['X1','X2'])
df.head()
```

```python
# onehot encoding

pd.get_dummies(df).shape
```

```python
df['X1'].nunique()
```

```python
df['X2'].nunique()
```

```python
## count for each one of the variable in X2

df.X2.value_counts().to_dict()
```

```python
## make a dictionary that maps each label to the counts

df_freq_map = df.X2.value_counts().to_dict()
```

```python
## we replace X2 labels in the dataset df

df.X2 = df.X2.map(df_freq_map)

df.head()
```
