## Data visualization
* It is a concept of graphical representation of data or information using visual elements like graphs, charts, maps.
* This representation helps us in understanding the patterns, trends and outliers in the data and it makes easily understandable and explainable.
* In data visualization, plot is a basic structure for the graphical representation of data.

### Plot
* Plot is a basic visualization element that helps to visualize the data. To visualize the data, required is figure and axes objects.
    * Figure: It is top-level container that acts as the window or page on which everything is drawn. It can contain multiple  independents sublplots, multiple axes, a title, a legend, etc.
    * Axes: The axes are the area on which data is plotted. it can have labels or ticks associated with it. There can be multiple Axes in a figure. But a given Axes object can only be in one figure.
    
### Components of plot
* Title: Name of the plot, by default, the title will be present at the top center of the plot
* Axis: This consists of the X and Y axis which provide the horizontal and vertical coordinates of the data points
* Labels: used to name the X and Y axis with appropriate names respectively
* Legend: set of key-value pairs that contains the list of plots and their respective labels.

## Matplotlib
* One of the most basic and popular python libraries used for data visualization
* It was developed for plotting capabilities of MATLAB
* **matplotlib.pyplot** is used for two-dimensional graphics in python programming.
* It can be used in power shell, scripts, web application servers and other graphical user interface toolkits.
* Matplotlib uses libraries such as NumPy as a base for underlying operations.
* Approaches to plotting in matplotlib:
    1. MATLAB way of plotting using matplotlib.pyplot, simple to use
    2. Object-oriented way of plotting for more control and customization

### Importing libraries

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
#creating two arrays
x=np.array([1,2,3,4,5])
y = x**2
```

  y: data on vertical axis

### Object oriented interface in Matplotlib
* Steps:
1. Creating a figure
2. Setting up the axes
3. Creating multiple plots using the same axes object
4. Setting up the title, label and legend for a plot

```python
#creating sample data for plotting
x = np.arange(1,16) 
y = x**2
z = (y/2)+40
```

Creating an empty canvas or figure, to plot the values and visualize the plot

```python
fig = plt.figure()
```

```python
ax = fig.add_axes([0,0,1,1])
ax.figure
```

#### Step 3:
Plotting a line on axis
* Syntax: ax.plot(x,y,color,lable)

```python
ax.plot(x,y,color="g",label='line1')
```

#### Step 4:
Plotting multiple lines on the axes

```python
ax.plot(x,z,color="r",label='line2')
```

#### Step 5:
Setting the title for the plot and the labels for the axes

#### Step 6:
Adding legend to the plot

```python
#Overall code for plotting
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="g",label='line1')
ax.plot(x,z,color="r",label='line2')
ax.set_title('Title of the plot',fontsize=15)
ax.set_xlabel('Label for X axis',fontsize=15)
ax.set_ylabel('Label for Y axis',fontsize=15)
ax.legend()
ax.figure
```

#### Creating multiple axes

```python
#creating an empty canvas figure
fig = plt.figure(figsize=(8,5))
#setting axes as [left,bottom,width,height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
#plotting the lines
ax1.plot(x,y,color="g",label='line1')
ax1.set_title('Title of the plot 1',fontsize=15)
ax1.set_xlabel('Label for X axis - plot 1',fontsize=15)
ax1.set_ylabel('Label for Y axis - plot 1',fontsize=15)
ax2.plot(x,z,color="r",label='line2')
ax2.set_title('Title of the plot 2',fontsize=15)
ax2.set_xlabel('Label for X axis - plot 2',fontsize=15)
ax2.set_ylabel('Label for Y axis - plot 2',fontsize=15)
ax1.legend()
ax2.legend()
```

#### subplots
* This method is used to create a common layout for multiple plots.
* m by n number of plots can be created using a subplot where m and n represent the number of rows and columns respectively.
* Syntax: plt.subplots(nrows,ncols)

```python
#initializing two axes for the subplot
fig, (ax1,ax2) = plt.subplots(1,2)

ax1.plot(x,y,color="g",label='line1')
ax1.set_title('Title of the plot 1',fontsize=15)
ax1.set_xlabel('Label for X axis - plot 1',fontsize=15)
ax1.set_ylabel('Label for Y axis - plot 1',fontsize=15)
ax2.plot(x,z,color="r",label='line2')
ax2.set_title('Title of the plot 2',fontsize=15)
ax2.set_xlabel('Label for X axis - plot 2',fontsize=15)
ax2.set_ylabel('Label for Y axis - plot 2',fontsize=15)
```

In this plt, the y label of the second plot is overlapping with the first plot, To avoid this, use **`fig.tight_layout()`**

```python
#initializing two axes for the subplot
fig, (ax1,ax2) = plt.subplots(1,2)

ax1.plot(x,y,color="g",label='line1')
ax1.set_title('Title of the plot 1',fontsize=15)
ax1.set_xlabel('Label for X axis - plot 1',fontsize=15)
ax1.set_ylabel('Label for Y axis - plot 1',fontsize=15)
ax2.plot(x,z,color="r",label='line2')
ax2.set_title('Title of the plot 2',fontsize=15)
ax2.set_xlabel('Label for X axis - plot 2',fontsize=15)
ax2.set_ylabel('Label for Y axis - plot 2',fontsize=15)
fig.tight_layout()
```

## Plotting with real time data

```python
#importing library
import pandas as pd

#importing df
df = pd.read_csv('./datasets/autompg1627540133031.csv')
df.info()
```

```python
#replacing the na values
df['horsepower'].fillna(df['horsepower'].mean(),inplace=True)
df.info()
```

### Boxplot
* Gives good indication of distribution of data about the median
* It is standardized way of displaying the distribution of data based on the five number summary

```python
#creating an empty canvas
fig = plt.figure(figsize=[6,6])

#setting the axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of mileage')

#plotting boxplot
ax.boxplot(df['mpg'])
fig.show()
```

There is no data for city mileage, but city mileage is 25% less than the averge mileage i.e. 'mpg'

```python
#processing data for city mileage
df['city_mileage'] = df['mpg']*0.75
df.head()
```

```python
#comparing average mpg and city mpg using boxplot
mpg_list = [df['mpg'],df['city_mileage']]
#creating an empty canvas
fig = plt.figure(figsize=[6,6])

#setting the axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of Average mpg vs city mpg')
ax.set_ylabel('Mileage per gallon')

#plotting boxplot
ax.boxplot(mpg_list)
fig.show()
```

### Scatter plot
* It uses dots or markers to represent a value on the axes.
* It can accept both quantitative and qualitative values.
* Used to identify correlation between variables
* Syntax: ax.scatter(x,y,marker)

```python
#creating an empty canvas
fig = plt.figure(figsize=[10,5])

#setting the axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Scatter plot of horsepower and mileage',fontsize=16)
ax.set_xlabel('Mileage per Gallon',fontsize=12)
ax.set_ylabel('Horsepower',fontsize=12)

#plotting scatter plot
ax.scatter(df['horsepower'],df['mpg'])
fig.show()
```

```python
#creating an empty canvas
fig = plt.figure(figsize=[10,5])

#setting the axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Scatter plot of horsepower and mileage based on origin',fontsize=16)
ax.set_xlabel('Mileage per Gallon',fontsize=12)
ax.set_ylabel('Horsepower',fontsize=12)
origin_list = df['origin'].unique()
colors = cm.jet(np.linspace(0,1,len(origin_list)))

#plotting scatter plot
for origin,color in zip(origin_list,colors):
    x=df[df['origin'] == origin]['horsepower']
    y=df[df['origin'] == origin]['mpg']
    plt.scatter(x,y,color=color,label=origin)
plt.legend
```

### Bar chart
* It is a graph with rectangular bars that usually compare different categories
* Each bar represents particular category
* Length of bar indicates the total number of values of items in the category
* can be plotted horizontally or vertically
* Syntax: ax.scatter(x,height,width,bottom,align) \
  x: data \
  height: height of bar \
  width: width of var, default 0.8 \
  bottom: Y coordinates of bar bases, default: 0 \
  align: alignment of the bars to x coordinates, default: center
  

```python
#creating a grouped df according to model year
grouped_df = df.groupby(['model_year']).count()[['name']]
grouped_df.reset_index(inplace=True)

#plotting the number of cars in each year
#creating a empty canvas
fig = plt.figure(figsize=[7,5])

#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title("number of cars in each year",fontsize=16)
ax.set_xlabel("Number of cars",fontsize=12)
ax.set_ylabel("Model Year",fontsize=12)

#plotting bar graph
ax.bar(grouped_df['model_year'],grouped_df['name'])
fig.show()
```

### Histogram
* Used for continuous data
* Each bar groups the numbers into intervals and the height of the bar is based on the number of values that fall into the corresponding intervals
* for frequency distribution
* Syntax: ax.hist(x,bins) \
  x: input values, single array or sequence of arrays \
  bins: int or sequence or str. if bins is integer, it defines the number of equal width bins in the range

```python
# creating histogram for horsepower
#creating a empty canvas
fig = plt.figure(figsize=[7,5])

#setting axes
ax = fig.add_axes([0,0,1,1])

#plotting histogram
ax.hist(df['horsepower'],bins=20)
ax.set_title('Distribution of horsepower',fontsize=17)
ax.set_xlabel('Frequency',fontsize=12)
ax.set_ylabel('Horsepower',fontsize=12)
fig.show()
```

### Pie chart
* Divides the entire data into distinct groups
* It consists of circle split into pies and each pie represents a group
* Size of pie proportional to the number of items in each group compared to others
* sum of pies always equal to 100%
* Syntax: ax.pie(x,labels) \
  x: wedge size, one-dimensional array \
  labels: sequence of strings providing the label for wedges


```python
pie_df = pd.DataFrame() #Creating sub df to plot the pie chart
pie_df['Count'] = df['origin'].value_counts() #getting the count of 'origin' and assigning to the df
pie_df = pie_df.reset_index() #re-arranging this index
pie_df.rename(columns={'index':'Country'},inplace=True) #re-naming the col
pie_df
```

```python
#creating a empty canvas
fig = plt.figure(figsize=[8,6])

#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Origin of the cars')

#plotting pie chart
ax.pie(pie_df['Count'],labels=pie_df['Country'])
fig.show()
```

#### Parameters of pie chart
* explode: To get an elevated view for the selected pie
* colors: To customize the colors for the plot
* autopct: To add the percentage of the distribution in the pie chart
* shadow: To add shadow to the plot
* startangle: To change the starting angle of the pie chart

```python
#creating a empty canvas
fig = plt.figure(figsize=[8,6])

#setting axes
ax = fig.add_axes([0,0,1,1])
explode = [0.05,0,0]
colors = ['yellowgreen','mediumaquamarine','khaki']
ax.set_title('Origin of the cars')

#plotting pie chart
ax.pie(pie_df['Count'],labels=pie_df['Country'],
      autopct='%0.1f%%',explode=explode,
      shadow=True,startangle=250,
      colors = colors)
fig.show()
```

### line chart
* It is drawn by interconnecting all data points using straight line segments in sequential order
* Used to analyze historic variations and trends in data
* Individual data points are chronologically connected to obtain the patterns and draw meaningful inferences

```python
#creating a df containing the mean values of mileage, horsepower and the weight of cars
line_df = df.groupby(by=['model_year'])
line_df = line_df.mean()
```

```python
#creating a empty canvas
fig = plt.figure(figsize=[8,6])

#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Mileage vs weight vs horsepower', fontsize=17)
ax.set_xlabel('Value',fontsize=12)
ax.set_ylabel('model year',fontsize=12)

#plotting line chart
ax.plot(line_df['mpg']/10,label='Mileage/10')
ax.plot(line_df['weight']/1000,label='Weight/1000')
ax.plot(line_df['horsepower']/100,label='horsepower/100')

ax.legend()
fig.show()
```

#### linestyle
* supports several line styles such as solid line, dashed line, dotted line, dashdot line, etc
* In addition to linestyle, the width of the line can be customized to get a unique plot

```python
#creating a empty canvas
fig = plt.figure(figsize=[8,6])

#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Mileage vs weight vs horsepower', fontsize=17)
ax.set_xlabel('Value',fontsize=12)
ax.set_ylabel('model year',fontsize=12)

#plotting line chart
ax.plot(line_df['mpg']/10,label='Mileage/10',linestyle='-',linewidth=2)
ax.plot(line_df['weight']/1000,label='Weight/1000',linestyle=':',linewidth=3)
ax.plot(line_df['horsepower']/100,label='horsepower/100',linestyle='--',linewidth=3)

ax.legend()
fig.show()
```

#### Text and annotate
* Text can also be added to describe the plot

```python
#creating a empty canvas
fig = plt.figure(figsize=[8,6])

#setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Mileage vs weight vs horsepower', fontsize=17)
ax.set_xlabel('Value',fontsize=12)
ax.set_ylabel('model year',fontsize=12)

#plotting line chart
ax.plot(line_df['mpg']/10,label='Mileage/10')
ax.plot(line_df['weight']/1000,label='Weight/1000')
ax.plot(line_df['horsepower']/100,label='horsepower/100')
plt.text(75.5,1.14,'Horsepower',
        va = 'center', #va -> vertical alignment
        rotation = 4, #angle
        bbox = dict(boxstyle = 'round',
                   facecolor='wheat',
                   alpha=0.3)) #background style

plt.text(75.5,2.25,'Mileage',
        va = 'center', #va -> vertical alignment
        rotation = 17, #angle
        bbox = dict(boxstyle = 'round',
                   facecolor='wheat',
                   alpha=0.3)) #background style

plt.text(76,3.15,'Weight',
        va = 'center', #va -> vertical alignment
        rotation = -13, #angle
        bbox = dict(boxstyle = 'round',
                   facecolor='wheat',
                   alpha=0.3)) #background style

plt.annotate('Highest Mileage',xy=(80.1,3.30),xytext=(80.5,3.38),
            arrowprops=dict(facecolor='wheat',shrink=0.01))
plt.annotate('Lowest Mileage',xy=(73.05,1.70),xytext=(73.5,1.69),
            arrowprops=dict(facecolor='wheat',shrink=0.01))


ax.legend()
fig.show()
```
