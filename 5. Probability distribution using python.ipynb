## Discrete Uniform Distribution
#### for calculating PMF, use randint.pmf function present in the stats module of Scipy python package

```python
#Discrete uniform distribution
from scipy.stats import randint

x = 1 #change the values to 2,3,4,5 and check
low = 1
high = 6 #total no. of possible outcomes +1

#Probability of occurrence of each outcome can be computed as follows:
drandint = randint.pmf(x,low,high)
print("when x=1, PMF is ", drandint)
```

#### PMF graph of discrete uniform distribution

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

low = 1
high = 6
x = np.arange(1,6,1)

#plot PMF
y1 = randint.pmf(x,low,high)

plt.bar(x,y1)
plt.xlabel("possible outcomes, x")
plt.ylabel("P(X=x)")
plt.title("The Discrete Uniform Ditribution - PMF")
plt.ylim(0,0.5)

plt.show()
```

#### for calculating CDF, use randint.cdf function present in the stats module of Scipy python package

```python
from scipy.stats import randint

x=1
low = 1
high = 6

prandint = randint.cdf(x,low,high)
print("when x = 1, CDF is ", prandint)
```

#### CDF graph of discrete uniform distribution

```python
low = 1
high = 6
x = np.arange(1,6,1)

#plot CDF
y2 = randint.cdf(x,low,high)
plt.bar(x,y2)
plt.xlabel("Possible outcomes, x")
plt.ylabel("P(X<=x)")
plt.title("CDF Discrete Uniform Distribution")
plt.ylim(0,1.1)

plt.show()
```

## Binomial Distribution

### Python functions
* PMF: binom.pmf(x,n,p)
* CDF: binom.cdf(x,n,p)
* Quantile function: binom.ppf(q,n,p) - The smallest value x s.t. F(x) >= p is defined as quantile where F is CDF
* Random generation: binom.rvs(n,size=10,p)

#### Generate random variables

```python
from scipy.stats import binom
x = 1 #no. of successes
n = 5 #no. of fixed trials
p = 0.2 #probability of success

rbinom = binom.rvs(n,p,size=10)
rbinom
```

#### PMF of binomial distribution

```python
x=2
pbinom = binom.pmf(x,n,p)
print("P(X=2) = ",round(pbinom,2))
```

#### PMF graph of binomial distribution

```python
n = 5
p = 0.2
x = np.arange(0,6,1)

#plot PMF
y1 = binom.pmf(x,n,p)

plt.bar(x,y1)
plt.xticks(x)
plt.xlabel('no. of successes x, in n trials')
plt.ylabel('P(X=x)')
plt.title('The Binomial Distribution PMF')

plt.show()
```

#### CDF of binomial distribution

```python
x = 2

cbinom = binom.cdf(x,n,p)
print("P(X<=2) = ", cbinom)
```

Since the quantile is the inverse of CDF, given the probability you can compute the number of failures that might occur until 2 successes are achieved as follows 

```python
q = binom.cdf(x,n,p)
qbinom = binom.ppf(q,n,p)
qbinom
```

#### CDF of binomial distribution

```python
n = 5
p = 0.2
x = np.arange(0,6,1)

#CDF
y2 = binom.cdf(x,n,p)

plt.bar(x,y2)
plt.xlabel("No. of successes x, in n trials")
plt.ylabel("P(X<=x)")
plt.title("Binomial Distribution CDF")

plt.show()
```

#### Ways to calculate normal probabilities

```python
from scipy.stats import norm
# by using z value
x = 84
mu = 95
sd = 11

z = (x-mu)/sd

pnorm = norm.cdf(z)
pnorm
```

```python
# without using z value
x = 84
mu = 95
sd = 11

pnorm = norm.cdf(x,mu,sd)
pnorm
```

#### To generate random variables

```python
mu = 0
sd = 2

x1 = norm.rvs(mu,sd,size=1000)

y1 = norm.pdf(x1,mu,sd)

plt.scatter(x1,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('The Normal distribution')
plt.show()
```

### Joint probabibility distribution plot

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

mu = np.array([1,2,3])
sigma = np.matrix([[20,15,0],[8,6,0],[0,0,80]])
data = np.random.multivariate_normal(mu,sigma,1000)

values = data.T

kde = stats.gaussian_kde(values)
density = kde(values)

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
x, y, z = values
ax.scatter(x, y, z, c=density)
plt.show()
```
